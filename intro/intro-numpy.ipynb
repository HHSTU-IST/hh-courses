# NumPy 入门

```python
# 导入numpy模块
import numpy as np
```

## 创建数组

```python
# 创建一维数组
a = np.array([1, 2, 3])
# 查看数组对象
a
```

```python
# 查看整数数组对象类型
a.dtype
```

```python
# 创建二维数组
c = np.array([[1, 2, 3], [4, 5, 6]])
c
```

```python
# 生成2×3的全0数组
np.zeros((2, 3))
```

```python
# 生成3×4的全1数组
np.ones((3, 4), dtype=np.int16)
```

```python
# 生成2×3的未初始化数组
np.empty([2, 3])
```

```python
# arange方法用于创建给定范围内的数组
np.arange(10, 30, 5)
```

```python
# 生成3×2的符合(0,1)均匀分布的随机数数组
np.random.rand(3, 2)
```

```python
# 生成0到2范围内长度为5的数组
np.random.randint(3, size=5)
```

```python
# 生成一组符合标准正态分布的随机数数组
np.random.randn(3)
```

## 数组切片

```python
# 创建一个一维数组
a = np.arange(10) ** 2
a
```

```python
# 获取数组的第3个元素
a[2]
```

```python
# 获取第2个到第4个数组元素
a[1:4]
```

```python
# 一维数组翻转
a[::-1]
```

```python
# 创建一个多维数组
b = np.random.random((3, 3))
b
```

```python
# 获取第2行第3列的数组元素
b[1, 2]
```

```python
# 获取第2列数据
b[:, 1]
```

```python
# 获取第3列前两行数据
b[:2, 2]
```

## 广播机制

```python
# 创建两个不同的数组
a = np.arange(4)
b = np.array([5, 10, 15, 20])
# 两个数组做减法运算
b - a
```

```python
# 数组的逻辑运算
b < 20
```

```python
# 数组求均值和方差
np.mean(b)
```

```python
# 数组求方差
np.var(b)
```

## 矩阵运算

```python
# 创建两个不同的数组
A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
# 矩阵元素乘积
A * B
```

```python
# 矩阵点乘
A.dot(B)
```

```python
# 矩阵求逆
np.linalg.inv(A)
```

```python
# 矩阵求行列式
np.linalg.det(A)
```

```python
# 创建一个3×4的数组
a = np.floor(10 * np.random.random((3, 4)))
a
```

```python
# 查看数组维度
a.shape
```

```python
# 数组展平
a.ravel()
```

```python
# 将数组变换为2×6数组
a.reshape(2, 6)
```

```python
# 求数组的转置
a.T
```

```python
a.T.shape
```

## 形式变换

```python
# -1维度表示NumPy会自动计算该维度
a.reshape(3, -1)
```

```python
# 按行合并代码清单1-7中的A数组和B数组
np.hstack((A, B))
```

```python
# 按列合并A数组和B数组
np.vstack((A, B))
```

```python
# 创建一个新数组
C = np.arange(16.0).reshape(4, 4)
C
```

```python
# 按水平方向将数组C切分为两个数组
np.hsplit(C, 2)
```

```python
# 按垂直方向将数组C切分为两个数组
np.vsplit(C, 2)
```
