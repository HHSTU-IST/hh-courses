```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 生成实验数据，0-10的50个点
x = np.linspace(0, 10, 50)
y1 = 2 * x + np.random.randn(50)  # 折线图数据
y2 = x**2 + np.random.randn(50) * 5  # 散点图数据
```

## 基本绘图

### 折线图

```python
_, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y1, color="#2E86AB", linewidth=2, label="data")
# 统一设置属性
ax.set(xlim=(0, 10), ylim=(-5, 25))
ax.legend()
```

### 散点图

```python
# 设置中文字体为宋体或黑体
plt.rcParams["font.sans-serif"] = ["SimSong"]
```

```python
_, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y2, color="#A23B72", s=30, alpha=0.7, label="二次数据")
ax.set(title="NumPy数据散点图", xlabel="X轴", ylabel="Y轴")
ax.legend()
```

### 坐标轴刻度变换

```python
_, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y2, color="#F18F01", linewidth=2)
# X轴线性刻度，Y轴对数刻度
ax.set(
    title="坐标轴刻度变换（Y轴对数刻度）",
    xlabel="X轴（线性）",
    ylabel="Y轴（对数）",
    yscale="log",  # 刻度变换
)
ax.grid(True, linestyle="--", alpha=0.5)
```

## 子图

```python
y3 = np.random.randint(1, 10, 5)  # 分类数据
categories = ["A", "B", "C", "D", "E"]
```

```python
# 创建子图
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), constrained_layout=1)

# 子图1
axs[0, 0].plot(x, y1, color="teal", linewidth=2, label="折线")
axs[0, 0].set(title="折线图", xlabel="X轴", ylabel="Y轴", xlim=(0, 10))
axs[0, 0].legend()
axs[0, 0].grid(alpha=0.3)

# 子图2
axs[0, 1].scatter(x, y2, color="orange", s=40, alpha=0.7, label="散点")
axs[0, 1].set(title="散点图", xlabel="X轴", ylabel="Y轴")
axs[0, 1].legend()

# 子图3
axs[1, 0].bar(categories, y3, color="purple", alpha=0.7)
axs[1, 0].set(title="柱状图", xlabel="类别", ylabel="数值")

# 子图4
axs[1, 1].plot(x, y2, color="red", linewidth=2)
axs[1, 1].set(
    title="Y轴对数刻度",
    xlabel="X轴",
    ylabel="Y轴(log)",
    yscale="log",  # 坐标轴刻度变换
)
axs[1, 1].grid(linestyle="--", alpha=0.4)
```
