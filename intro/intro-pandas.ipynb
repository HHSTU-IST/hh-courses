```python
import pandas as pd
```

## 文件读写

### 创建测试数据

```python
data = {
    "姓名": ["张三", "李四", "王五", "赵六"],
    "年龄": [20, 21, 19, 22],
    "成绩": [85.5, 92.0, 78.5, 88.0],
}
student = pd.DataFrame(data)
student.head()
```

### 保存文件

```python
student.to_csv("student.csv", index=False, encoding="utf-8-sig")
```

### 读取CSV文件

```python
student = pd.read_csv("student.csv", encoding="utf-8-sig")
student.head(2)
```

```python
# 数据集形状
student.shape
```

## 数据切片

### 列切片

```python
single_col = student["姓名"]
single_col
```

```python
multi_col = student[["姓名", "成绩"]]
multi_col
```

### 行切片

```python
# 行标签1-2
loc_slice = student.loc[1:2, :]
loc_slice
```

```python
# 行0-1，列1-2
iloc_slice = student.iloc[0:2, 1:3]
iloc_slice
```

### 条件切片

```python
cond_slice = student[student["成绩"] > 85]
cond_slice
```
